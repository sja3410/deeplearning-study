# 5-2 과대적합과 과소적합을 알아봅니다.

#학습 곡선을 통해 과대적합과 과소적합을 알아봅니다. 
과대 적합 : 훈련 세트에서는 좋은 성능을 내지만 검증 세트에서는 낮은 성능을 내는 경우
과소 적합 : 훈련 세트와 검증 세트의 성능에는 차이가 크지 않지만 모두 낮은 성능을 나타내는 경우

과대적합된 모델을 '분산이 크다'라고 한다. 
-과대적합의 주요 원인은 훈련 세트에 충분히 다양한 패턴의 샘플이 포함되지 않은 경우이다.
-훈련 샘플을 더 많이 모을 수 없는 상황에는 모델이 훈련 세트에 집착하지 않도록 가중치를 제한할 수 있다.

과소적합된 모델을 '편향이 크다'라고 한다.
-복잡도가 더 높은 모델을 사용하거나 가중치의 규제를 완화함으로써 해결 할 수 있다.

<에포크와 손실 함수의 그래프로 과대적합과 과소적합 분석하기>
-에포크에 대한 손실 함수의 그래프를 이용하여 과대적합과 과소적합을 분석하기도 한다.
-p132 에포크에 대한 검증 세트의 손실 함수와 훈련 세트의 손실함수가 있다.
-최적점 전까지의 에포크에서는 과소적합이 일어난다. --> 정확성이 떨어진다.
-최적점 후에서의 에포크부터는 과대적합이 일어난다. --> 훈련 세트에서는 에포크가 진행이 될수록 정확성을 높아지는 반면에 훈련이 훈련 세트에만 취우쳐서 진행 되므로 검증 세트에서는 더욱 정확도가 떨어지게 된다.

<모델 복잡도와 손실 함수의 그래프로 과대적합과 과소적합 분석하기>
-모델 복잡도 : 모델이 가진 학습 가능한 가중치 개수
-훈련 세트에서만 잘 맞는 형태로 만들어지면 훈련 세트에서만 좋은 성능을 내기 때문에 위의 에포크로 분석하는 것과 같은 그래프를 띈다.

#적절한 편향-분산 트레이드오프를 선택합니다.
-과소적합된 모델과 과대적합된 모델 사이의 관계를 '편향-분산 트레이드오프'라고 부른다. 
-트레이드 오프는 편향이 커지면 분산이 적어지고, 분산이 커지면 편향이 적어지기때문에 
-우리는 그 중간(최적)을 선택해야한다.--> 이것을 '적절한 편향-분산 트레이드오프를 선택했다.'라고한다.
